#Data Replacement and Cleaning

**Back Ups**

```python
milktea_df = acquiring_milktea_data.copy()
```

**List of Function Used**


*   Creating a new dataframe from existing dataframe function






```python
class NewDataFrame:
  def __init__(self, col_name, target_val):
    self.col_name = col_name
    self.target_val = target_val

  def make_df(self):
    convert = milktea_df.loc[milktea_df[self.col_name]==self.target_val]
    df=pd.DataFrame(convert)
    ## adjusting columns
    col_to_list = df.columns.tolist()
    index = col_to_list.index(self.col_name)
    new_pos_col = col_to_list[index:]+col_to_list[:index]
    df = df[new_pos_col]
    return df
```



*   Renaming a an existing column names function




```python
## For changing Multiple Column names 
def rename_cols(old_col_names, new_col_names):
  for i, old_col_name in enumerate(old_col_names):
    new_col_name = new_col_names[i]
    rename_col(old_col_name,new_col_name)

## For changing Single Column name
def rename_col(old_col_name, new_col_name):
  return milktea_df.rename(columns = {old_col_name:new_col_name}, inplace = True)
```



*   Find and replace function



```python
def find_list_replace(col_name, find_vals, replace_val):
  for find_val in find_vals:
    find_replace(col_name, find_val, replace_val)

def find_replace(col_name, find_val, replace_val):
  index = milktea_df.index[milktea_df[col_name]==find_val].tolist()
  milktea_df.loc[index,col_name] = replace_val
  return milktea_df.loc[index,col_name]
```



*   Deleting an entire column function



```python
def delete_col(col_name):
  return milktea_df.drop(labels=col_name, axis = 1)
```



*   Renaming Column Names
