#Data Replacement and Cleaning


*   Renaming Columns
*   Find and Replace Values
*   Deleting Irrelevant Column





*   Renaming a column names function
*   Find and replace function
*   Deleting an entire column function
*   Renaming Column Names



```python
class CleanData:
    def __init__(self,data, target_vals, new_vals):
        self.data = data
        self.target_vals = target_vals
        self.new_vals = new_vals

    def create_df(self):
        pass

    def update_col(self):
        for i, old_col_name in enumerate(self.target_vals):
            new_col_name = self.new_vals[i]
            self.change_col(self.data,old_col_name,new_col_name)
    
    def change_col(self, data, old_col_name, new_col_name):
        return data.rename(columns = {old_col_name:new_col_name}, inplace = True)
    
    def update_val(self):
        return self.data.replace(to_replace = self.target_vals, value = self.new_vals)

    def remove_space(self):
        self.data[self.target_vals] = self.data[self.target_vals].str.strip()
        return self.data

    def delete_col(self):
        return self.data.drop(labels = self.target_vals, axis = 1)
```

*   Creating a new dataframe function






```python
class NewDataFrame:
  def __init__(self, col_name, target_val):
    self.col_name = col_name
    self.target_val = target_val

  def make_df(self):
    convert = milktea_df.loc[milktea_df[self.col_name]==self.target_val]
    df=pd.DataFrame(convert)
    ## adjusting columns
    col_to_list = df.columns.tolist()
    index = col_to_list.index(self.col_name)
    new_pos_col = col_to_list[index:]+col_to_list[:index]
    df = df[new_pos_col]
    return df
```

# Data Transformation and Visualization

**Function for Plotting**

```python
class Plot:
    def __init__(self,val_x, val_y,val_y2,labelx, labely, name):
        self.val_x = val_x        
        self.val_y = val_y
        self.val_y2 = val_y2
        self.labelx = labelx
        self.labely = labely
        self.name = name
        
    def bar_graph(self):
        sns.set_style("darkgrid")
        plt.figure(figsize= (20, 10))
        plt.xticks(rotation = 80, horizontalalignment = "center")
        
        plt.bar(self.val_x, self.val_y, width = 0.5,color=(0.2, 0.4, 0.6, 0.6))
        plt.plot(self.val_x, self.val_y2, color = 'red')
        plt.title(self.name)
        plt.xlabel(self.labelx)
        plt.ylabel(self.labely)

    def bar_graph2(self):
        sns.set_style("darkgrid")
        plt.figure(figsize= (20, 10))
        plt.xticks(rotation = 80, horizontalalignment = "center")
        
        plt.bar(self.val_x, self.val_y, width = 0.5,color=(0.2, 0.4, 0.6, 0.6))
        plt.plot(self.val_x, self.val_y, 's-b')
        plt.title(self.name)
        plt.xlabel(self.labelx)
        plt.ylabel(self.labely)

    def line_graph(self):
        sns.set_style("darkgrid")
        plt.figure(figsize= (20, 10))
        plt.xticks(rotation = 80, horizontalalignment = "center")

        plt.plot(self.val_x, self.val_y, color = 'red', marker = 'o')
        plt.title(self.name)
        plt.xlabel(self.labelx)
        plt.ylabel(self.labely)
```
